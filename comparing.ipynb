```python
import statistics
import matplotlib.pyplot as plt
import pandas as pd

# List to store all traces
allTraces = []

#word-unroll-loop
#forceInline
#deleteIO
#precomputedConstants
#sample_change

en_f=['word-unroll-loop_en_energy.csv','forceInline_en_energy.csv','deleteIO_en_energy.csv',
     'precomputedConstants_en_energy.csv','sample_change_en_energy.csv']
de_f=['word-unroll-loop_de_energy.csv','forceInline_de_energy.csv','deleteIO_de_energy.csv',
     'precomputedConstants_de_energy.csv','sample_change_de_energy.csv']

# List of CSV file names
csv_files = de_f  # Add your file names here

# Loop through each CSV file
for file in csv_files:
    # Read CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # Extract data from DataFrame as a list
    #trace_data = df[0].tolist()
    
    # Append data to allTraces list
    #allTraces.append(trace_data)
    
    for column in df.columns:
        trace_data = df[column].tolist()
        allTraces.append(trace_data)

# Print allTraces
print(allTraces)


```

```python
len(allTraces)
```

```python
# allTraces=[[], [], []]
nameTraces=['word-unroll-loop','forceInline','deleteIO',
     'precomputedConstants','original']

avgPower=[]
stdPower=[]

for i in range(len(allTraces)):
    avgPower.append(statistics.mean(allTraces[i]))
    stdPower.append(statistics.stdev(allTraces[i]))

plt.bar(nameTraces, avgPower)
plt.ylim(0, max(avgPower) + 2)
plt.title('comparing different varients')
for i, value in enumerate(avgPower):
    plt.text(i, value + 1, format(value, '.4f'), ha='center', va='bottom')
plt.xlabel('varients')
plt.ylabel('average')
plt.xticks(rotation=45)

plt.show()
```
